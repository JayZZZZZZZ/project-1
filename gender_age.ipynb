```python
import pandas as pd
import numpy as np
```

```python
#import column info
df_gender_age_info = pd.read_csv("gender_age_metadata.csv") 
df_gender_age_info
```

```python
#create filter for desired columns
gender_age_info_rows = ["DP05_0001E", "DP05_0003E", "DP05_0004E", "DP05_0009E", "DP05_0010E", "DP05_0011E", "DP05_0012E", "DP05_0013E", "DP05_0014E", "DP05_0027E", "DP05_0028E"]
```

```python
#filter for desired columns
df_gender_age_info = df_gender_age_info[df_gender_age_info["GEO_ID"].isin(gender_age_info_rows)]
df_gender_age_info
```

```python
#clean column names
df_gender_age_info.loc[:, ('id')] = df_gender_age_info.id.map(lambda x: x.lstrip("Estimate!!SEX AND AGE!!"))
df_gender_age_info.loc[:, ('id')] = df_gender_age_info.id.replace("!!", " - ", regex=True)
df_gender_age_info
```

```python
#create dictionary of new column names
df_gender_age_column_key = df_gender_age_info.set_index('GEO_ID').T.to_dict('records')
df_gender_age_column_key = df_gender_age_column_key[0]
df_gender_age_column_key
```

```python
#create lists of columns
df_gender_age_numeric_columns = df_gender_age_info.GEO_ID.tolist()
df_gender_age_columns = []
for i in df_gender_age_numeric_columns:
    df_gender_age_columns.append(i)
df_gender_age_columns.append("GEO_ID")
df_gender_age_columns.append("NAME")
print(df_gender_age_numeric_columns)
print(df_gender_age_columns)
```

```python
#import census age_gender data, with only desired columns
df_gender_age = pd.read_csv("gender_age_data_with_overlays.csv", usecols=df_gender_age_columns)
df_gender_age
```

```python
#remove unwanted first row
df_gender_age = df_gender_age.iloc[1:] 
df_gender_age
```

```python
df_gender_age.info()
```

```python
#replace dashes in dataset with zeros
df_gender_age = df_gender_age.replace("-", 0)
```

```python
#change quantitative data types in dataframe to numeric
df_gender_age[df_gender_age_numeric_columns] = df_gender_age[df_gender_age_numeric_columns].apply(pd.to_numeric)
```

```python
#rename columns
df_gender_age = df_gender_age.rename(columns=df_gender_age_column_key)
df_gender_age
```
