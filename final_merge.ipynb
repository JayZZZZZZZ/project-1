```python
import pandas as pd
import numpy as np
import pickle
```

```python
key = pd.read_pickle("station_key.pkl")
key.rename(columns={"matches":"name"}, inplace = True)
key
```

```python
mta = pd.read_pickle("mta_daily_data_v1.pkl")
mta
```

```python
stations_w_data = pd.read_csv("https://raw.githubusercontent.com/dunleavyjason/project-1/Alexander/GeoSpatial%20Data/Stations_W_Data.csv")
stations_w_data
```

```python
geopandas_merged = pd.merge(stations_w_data, key, on='name', how='left')
geopandas_merged
```

```python
geopandas_merged.drop_duplicates(subset=["STATION"])
```

```python
final_merge = pd.merge(mta, geopandas_merged, on='STATION', how='left')
final_merge.sample(50)
```

```python
final_merge = final_merge[final_merge['STATION_CLEAN'].notna()]
final_merge
```

```python
final_merge.drop(columns =["Unnamed: 0", "STATION_CLEAN"], inplace = True)
final_merge.info()
```
