```python
import pandas as pd
import numpy as np
import pickle
```

```python
with open('df_education.pkl', 'rb') as read_file:
    df_education = pickle.load(read_file)
```

```python
df_education.set_index('GEO_ID')
```

```python
with open('df_gender_age.pkl', 'rb') as read_file:
    df_gender_age = pickle.load(read_file)
```

```python
df_gender_age.set_index('GEO_ID')
```

```python
df_census = df_education.join(df_gender_age, how = "outer", rsuffix = 'r')
df_census
```

```python
del df_census['GEO_IDr']
del df_census['NAMEr']
df_census
```

```python
df_census["Female Ratio"]=100/(df_census["Total population - Sex ratio (males per 100 females)"]+100)
```

```python
df_census["Total Female Population - 25 years and over"] = df_census["Total - AGE BY EDUCATIONAL ATTAINMENT - Population 25 years and over"]*df_census["Female Ratio"]
```

```python
df_census
```

```python
df_census["GEO_ID"]=df_census["GEO_ID"].str[9:]
df_census
```

```python
df_census.to_pickle("df_census.pkl")
```
