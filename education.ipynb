```python
import pandas as pd
import numpy as np
```

```python
#import column info
df_education_info = pd.read_csv("education_metadata.csv") 
df_education_info
```

```python
#create filter for desired columns
education_rows = ["S1501_C01_006E", "S1501_C01_015E","S1501_C05_012E","S1501_C01_059E","S1501_C05_063E","S1501_C05_013E",'S1501_C05_064E']
```

```python
#filter for desired columns
df_education_info = df_education_info[df_education_info["GEO_ID"].isin(education_rows)]
df_education_info
```

```python
#clean column names
df_education_info['id'] = df_education_info.id.map(lambda x: x.lstrip("Estimate!!"))
df_education_info['id'] = df_education_info.id.replace("!!", " - ", regex=True)
df_education_info
```

```python
#create dictionary of new column names
df_education_column_key = df_education_info.set_index('GEO_ID').T.to_dict('records')
df_education_column_key = df_education_column_key[0]
df_education_column_key
```

```python
#create lists of columns
df_education_numeric_columns = df_education_info.GEO_ID.tolist()
df_education_columns = []
for i in df_education_numeric_columns:
    df_education_columns.append(i)
df_education_columns.append("GEO_ID")
df_education_columns.append("NAME")
print(df_education_numeric_columns)
print(df_education_columns)
```

```python
#import census age_gender data, with only desired columns
df_education = pd.read_csv("education_data_with_overlays.csv", usecols=df_education_columns)
df_education
```

```python
df_education = df_education.iloc[1:] 
df_education
```

```python
df_education.info()
```

```python
#replace dashes in dataset with zeros
df_education = df_education.replace("-", 0)
```

```python
df_education.loc[120, ("S1501_C05_064E")] = 2500
df_education.loc[121, ("S1501_C05_064E")] = 2500
```

```python
#change quantitative data types in dataframe to numeric
df_education[df_education_numeric_columns] = df_education[df_education_numeric_columns].apply(pd.to_numeric)
```

```python
#rename columns
df_education = df_education.rename(columns=df_education_column_key)
df_education
```
